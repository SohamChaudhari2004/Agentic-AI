```python
# read docs
def read_docs(directory_path):
    loader = PyPDFDirectoryLoader(directory_path)
    documents = loader.load()
    return documents
```

```python
import pprint

docs = read_docs("./")
```

```python
docs
```

```python
def chunk_docs(documents, chunk_size=8000, chunk_overlap=100):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
    )
    chunks = text_splitter.split_documents(documents)
    return chunks
```

```python
documents = chunk_docs(docs)
```
